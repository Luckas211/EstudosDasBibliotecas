# Aula 07 - Otimizando leitura com Tesseract (PSM)

O Tesseract OCR (o robô leitor) é muito esperto, mas às vezes ele é um pouco cego sobre a *estrutura* do papel.

Por exemplo: se ele olha para uma tabela do Excel, ele pode se perder lendo a primeira palavra de cada coluna, ignorando o espaço entre elas. Para consertar isso, nós ajudamos o robô dizendo a ele qual é a estrutura do documento, usando os chamados **PSM (Page Segmentation Modes)**.

## O que são os PSMs?
São regras de como ler a página. Os mais usados no dia a dia:
*   **PSM 3:** (O Padrão) Tenta entender que a página tem texto comum (blocos de parágrafos).
*   **PSM 6:** Assume que é UM ÚNICO bloco uniforme de texto. Ótimo para recibos contínuos.
*   **PSM 11:** Procura o máximo possível de pequenos textos espalhados, ignorando a ordem. (Muitas vezes útil para ler os dados dentro das 'caixinhas' de uma tabela ou formulário).
