# Aula01 - Introducao a expressoes regulares

O modulo `re` permite procurar padroes dentro de um texto.

Pense assim:

- string comum: procura texto exato
- expressao regular: procura um formato

Exemplo de formato:

- um numero
- tres letras seguidas
- uma data como `21/04/2026`


## Funcao: `re.search(padrao, texto)`

Esta funcao procura o **primeiro ponto** do texto que combina com o padrao.

Parametros:

- `padrao`: a regra que voce quer buscar
- `texto`: a string onde a busca sera feita

Retorno:

- um objeto `Match` se encontrar
- `None` se nao encontrar


```python
import re  # Importa o modulo re, que contem funcoes para expressoes regulares.

texto = "O pedido 458 foi enviado."  # Cria uma string simples contendo uma frase com um numero.
padrao = r"\d+"  # Define um padrao que significa um ou mais digitos numericos.

resultado = re.search(padrao, texto)  # Procura a primeira sequencia de digitos dentro do texto.

print(resultado)  # Mostra o objeto Match para voce entender que houve uma correspondencia.

if resultado:  # Verifica se o resultado nao e None antes de acessar seus dados internos.
    print(resultado.group())  # Mostra o trecho exato do texto que combinou com o padrao.
```

## O que significa `\d+`

- `\d` significa "um digito"
- `+` significa "uma ou mais repeticoes"

Entao `\d+` quer dizer:
"encontre um ou mais numeros seguidos"


```python
import re  # Importa o modulo re para um segundo exemplo de busca.

texto = "Meu CEP e 12345-678."  # Cria um texto com um CEP no formato brasileiro.
padrao_cep = r"\d{5}-\d{3}"  # Define um padrao com cinco digitos, um hifen e tres digitos.

cep_encontrado = re.search(padrao_cep, texto)  # Procura no texto um trecho que siga o formato do CEP.

if cep_encontrado:  # Confirma que o trecho foi localizado.
    print(cep_encontrado.group())  # Mostra apenas o valor que casou com a expressao regular.
```

## Resumo

Hoje voce viu:

- `re.search()`
- o objeto `Match`
- o metodo `.group()`
- simbolos como `\d`, `+` e `{5}`
