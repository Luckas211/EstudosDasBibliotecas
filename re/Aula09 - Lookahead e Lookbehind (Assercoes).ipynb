# Aula 09 - Lookahead e Lookbehind (As Asserções!)

Chegamos na magia avançada das Expressões Regulares (Regex). 

Imagine a seguinte missão: 'Quero extrair o preço da frase, mas **SOMENTE** se o preço estiver **antes** da palavra *Reais*, e eu **NÃO** quero capturar a palavra Reais, só o número'.

Se usarmos o padrão simples `\d+ Reais`, ele captura '50 Reais'. Mas nós só queremos '50'. 
Para isso usamos **Lookarounds** (olhar ao redor). O Regex verifica a existência das palavras, mas não as inclui no recorte final.

```python
import re
```

## 1. Lookahead Positivo: `(?=...)` (Olhe para frente)
Procura algo que está LOGO APÓS nossa busca principal.

```python

import re
frase_1 = 'Eu tenho 5 gatos e custou 50 reais comprar a ração.'

# Queremos o número (\d+) seguido pela palavra 'reais', mas não incluí-la.
padrao_lookahead = r'\d+(?= reais)'

# Achou o 50, mas perceba que ignorou o '5' (porque não tem 'reais' depois dele) 
# e recortou SOMENTE o número 50.
resultado_1 = re.findall(padrao_lookahead, frase_1)
print('Número antes de Reais:', resultado_1)
```

## 2. Lookbehind Positivo: `(?<=...)` (Olhe para trás)
Procura algo que está LOGO ANTES da nossa busca principal.

```python
frase_2 = 'Meu produto custa R$250, mas o frete é R$15.'

# Queremos números (\d+) que venham EXATAMENTE APÓS 'R$'.
# Obs: No regex, o cifrão ($) é caractere especial de fim de linha, então colocamos barra (\$) pra dizer que é literal.
padrao_lookbehind = r'(?<=R\$)\d+'

resultado_2 = re.findall(padrao_lookbehind, frase_2)
print('Valores em Reais extraídos:', resultado_2)
```
