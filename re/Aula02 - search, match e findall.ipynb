# Aula02 - search, match e findall

Agora vamos comparar funcoes importantes do modulo `re`.

- `re.search()` procura em qualquer parte do texto
- `re.match()` procura apenas no inicio
- `re.fullmatch()` exige que o texto inteiro siga o padrao
- `re.findall()` devolve todas as ocorrencias


```python
import re  # Importa o modulo re para comparar formas diferentes de busca.

texto = "Produto 123 cadastrado"  # Cria um texto de exemplo para as verificacoes.
padrao = r"Produto"  # Define um padrao literal simples com a palavra Produto.

print(re.search(padrao, texto))  # Procura a palavra em qualquer posicao do texto.
print(re.match(padrao, texto))  # Procura a palavra apenas no comeco do texto.
print(re.fullmatch(padrao, texto))  # Exige que o texto inteiro seja exatamente Produto.
```

## Funcao: `re.findall()`

`findall` e muito util quando voce quer pegar todas as ocorrencias.

Retorno:

- uma lista com tudo o que foi encontrado


```python
import re  # Importa o modulo re para localizar varias ocorrencias no mesmo texto.

texto = "Os numeros sao 10, 20 e 30."  # Cria um texto contendo tres numeros.
padrao = r"\d+"  # Define um padrao que encontra um ou mais digitos.

numeros = re.findall(padrao, texto)  # Procura todas as sequencias numericas existentes no texto.

print(numeros)  # Mostra a lista com todos os numeros encontrados.
print(type(numeros))  # Mostra o tipo do retorno para reforcar que findall devolve lista.
```

## Funcao: `re.finditer()`

`finditer` e parecida com `findall`, mas devolve objetos `Match`.
Isso permite descobrir:

- o texto encontrado
- a posicao inicial
- a posicao final


```python
import re  # Importa o modulo re para percorrer correspondencias com mais detalhes.

texto = "A sala 12 fica ao lado da sala 25."  # Cria um texto com mais de um numero.
padrao = r"\d+"  # Define o padrao que encontra blocos numericos.

for correspondencia in re.finditer(padrao, texto):  # Percorre cada objeto Match encontrado no texto.
    print(correspondencia.group())  # Mostra o valor textual capturado pela correspondencia atual.
    print(correspondencia.start())  # Mostra o indice inicial em que a correspondencia comeca.
    print(correspondencia.end())  # Mostra o indice final da correspondencia.
    print("-" * 30)  # Separa visualmente uma correspondencia da outra.
```
