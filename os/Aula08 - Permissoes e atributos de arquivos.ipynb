# Aula 08 - Permissões e Atributos de Arquivos

Bem-vindo à Aula 08! Hoje vamos falar sobre os 'bastidores' dos arquivos no seu computador.

Imagine que cada arquivo tem uma 'identidade' e um 'crachá de segurança'. A identidade diz qual o tamanho dele, quando foi criado e modificado (são os **atributos**). Já o crachá de segurança diz quem pode ler, quem pode escrever e executar (são as **permissões**).

Para um iniciante, isso é super útil: imagine que você fez um script que cria um arquivo importante e você quer protegê-lo para que ninguém apague o conteúdo sem querer (tornando-o 'Somente Leitura'). É isso que vamos aprender aqui!

## 1. Lendo a 'Identidade' do Arquivo (os.stat)

A função `os.stat` é como um scanner de raio-x. Ela olha para o arquivo e nos devolve um monte de informações (metadados).

```python
import os
import stat # Nos ajuda a lidar com as permissões de forma legível
import time # Converte tempos do formato de máquina (segundos) para datas humanas

arquivo_teste = 'meu_arquivo_secreto.txt'
with open(arquivo_teste, 'w') as f:
    f.write('Este é um documento muito importante!')

# Passando o 'raio-x'
informacoes = os.stat(arquivo_teste)

# st_size = Tamanho em bytes
print('Tamanho do arquivo:', informacoes.st_size, 'bytes')
# Usamos o time.ctime() para converter segundos em data legível
print('Última vez modificado:', time.ctime(informacoes.st_mtime))
```

## 2. Alterando o 'Crachá de Segurança' (os.chmod)

O `chmod` (Change Mode) altera o comportamento. Podemos usar:
* `stat.S_IREAD`: Permissão de LEITURA (Somente Leitura).
* `stat.S_IWRITE`: Permissão de ESCRITA.

```python
# Vamos tornar o arquivo Somente Leitura!
os.chmod(arquivo_teste, stat.S_IREAD)
print('Arquivo agora é Somente Leitura!')

# Se tentar escrever agora, dará Erro de Permissão!
# Para voltar ao normal, damos a permissão de ESCRITA novamente
os.chmod(arquivo_teste, stat.S_IWRITE)
print('Arquivo pode ser modificado novamente!')

# Apagando arquivo de teste
os.remove(arquivo_teste)
```
