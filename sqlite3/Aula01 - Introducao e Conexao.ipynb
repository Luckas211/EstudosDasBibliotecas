# Aula 01 - Introdução e Conexão com SQLite
O SQLite é um banco de dados relacional leve, baseado em arquivo, que já vem embutido no Python na biblioteca `sqlite3`. Ele é ideal para pequenos projetos, aprendizado, prototipagem e até para muitas aplicações em produção.


## Importando e Conectando
Para usar, basta importar a biblioteca e usar o método `.connect()`. Se o arquivo de banco de dados não existir no local especificado, ele será criado automaticamente.


```
import sqlite3

# Conecta ao banco de dados (cria o arquivo meu_banco.db se não existir)
conn = sqlite3.connect('meu_banco.db')

print('Conexão estabelecida com sucesso!')

# Sempre feche a conexão quando terminar para liberar os recursos
conn.close()
```
