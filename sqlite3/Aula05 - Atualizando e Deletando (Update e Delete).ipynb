# Aula 05 - Atualizando (Update) e Deletando (Delete)
O **UPDATE** modifica registros que já existem.
O **DELETE** remove registros do banco.

⚠️ **MUITO CUIDADO**: Sempre use a cláusula `WHERE` (ONDE). Se você esquecer o `WHERE`, você vai atualizar ou deletar **TODOS** os registros da sua tabela de uma só vez!


## Atualizando dados


```
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Vamos atualizar a idade do Carlos para 31 anos
cursor.execute("UPDATE usuarios SET idade = ? WHERE nome = ?", (31, 'Carlos'))
conn.commit()

print("Idade do Carlos atualizada!")
```

## Deletando dados


```
# Vamos deletar a Ana do banco de dados
cursor.execute("DELETE FROM usuarios WHERE nome = ?", ('Ana',))
conn.commit()

print("Usuário Ana deletado com sucesso!")
conn.close()
```
