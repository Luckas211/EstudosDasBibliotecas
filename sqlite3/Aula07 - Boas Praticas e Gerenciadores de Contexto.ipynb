# Aula 07 - Boas Práticas e Gerenciadores de Contexto
Uma forma muito 'Pythonica' e segura de trabalhar com bancos de dados é usando o `with` (Gerenciador de Contexto).

Quando usamos o `with`, o próprio Python cuida de fazer o `.commit()` se o código rodar sem erros. E se der algum erro durante o processo, ele faz o `.rollback()` (desfaz as alterações), garantindo a integridade dos dados!


```
import sqlite3

# Usando with para gerenciar a transação
with sqlite3.connect('meu_banco.db') as conn:
    cursor = conn.cursor()
    
    cursor.execute("INSERT INTO usuarios (nome, idade, email) VALUES (?, ?, ?)", ('Eduardo', 28, 'edu@email.com'))
    
    # Perceba que não precisamos usar conn.commit()!
    print("Eduardo inserido de forma segura.")
```

## Transformando Resultados em Dicionários
Por padrão, o SQLite retorna tuplas `(1, 'Carlos', ...)`. Isso pode ser chato de ler. Podemos configurar o banco para retornar **Dicionários**, onde podemos acessar via `linha['nome']`.


```
def dict_factory(cursor, row):
    d = {}
    for idx, col in enumerate(cursor.description):
        d[col[0]] = row[idx]
    return d

with sqlite3.connect('meu_banco.db') as conn:
    # Muda o formato de saída das linhas
    conn.row_factory = dict_factory
    
    cursor = conn.cursor()
    cursor.execute("SELECT nome, email FROM usuarios")
    
    resultados = cursor.fetchall()
    
    for usuario in resultados:
        print(f"Nome: {usuario['nome']} | E-mail: {usuario['email']}")
```
