# Aula 02 - Criando Tabelas (Create)
Para executar comandos SQL no banco de dados, precisamos de um objeto chamado `cursor`. Com ele, podemos executar a instrução SQL `CREATE TABLE`.


```
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Criando uma tabela de usuários
# Usamos IF NOT EXISTS para evitar que o código dê erro caso a tabela já tenha sido criada antes
cursor.execute('''
CREATE TABLE IF NOT EXISTS usuarios (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL,
    idade INTEGER,
    email TEXT UNIQUE
)
''')

# Confirma as alterações (Obrigatório para operações que modificam o banco)
conn.commit()

print('Tabela criada com sucesso!')
conn.close()
```
