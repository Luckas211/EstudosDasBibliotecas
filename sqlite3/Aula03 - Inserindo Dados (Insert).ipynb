# Aula 03 - Inserindo Dados (Insert)
Agora que temos uma tabela, vamos inserir dados usando a instrução `INSERT INTO`.

⚠️ **ATENÇÃO:** Nunca utilize concatenação de strings básica (`f-strings` ou `+`) para inserir dados no SQL. Isso abre brechas para ataques de **SQL Injection**. Use sempre os placeholders `?`.


## Inserindo um único registro


```
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# O '?' é o nosso placeholder. Os valores são passados numa tupla no segundo argumento.
cursor.execute("INSERT INTO usuarios (nome, idade, email) VALUES (?, ?, ?)", ('Ana', 25, 'ana@email.com'))

conn.commit() # Salva a transação no banco de dados
print('Registro da Ana inserido com sucesso!')
```

## Inserindo múltiplos registros de uma vez (`executemany`)


```
usuarios_lista = [
    ('Carlos', 30, 'carlos@email.com'),
    ('Beatriz', 22, 'bia@email.com'),
    ('Daniel', 40, 'daniel@email.com')
]

# O executemany aplica a mesma instrução SQL para cada item da lista
cursor.executemany("INSERT INTO usuarios (nome, idade, email) VALUES (?, ?, ?)", usuarios_lista)
conn.commit()

print('Múltiplos registros inseridos com sucesso!')
conn.close()
```
