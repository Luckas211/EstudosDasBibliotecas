# Aula 04 - Lendo Dados (Read)
Para buscar os dados armazenados, usamos a instrução `SELECT`.
Depois de executar a busca com o `cursor`, usamos os métodos:
- `fetchone()`: Retorna apenas o primeiro resultado encontrado.
- `fetchall()`: Retorna uma lista com todos os resultados encontrados.


```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Seleciona TODAS as colunas (*) da tabela usuarios
cursor.execute("SELECT * FROM usuarios")

# Pega os resultados
resultados = cursor.fetchall()

print("Todos os usuários cadastrados:")
for linha in resultados:
    # Cada linha é retornada como uma tupla (id, nome, idade, email)
    print(linha)

conn.close()
```
