# Aula 06 - Consultas Avançadas e Filtros
O `SELECT` é muito poderoso. Podemos usar filtros condicionais e ordenadores para trazer os dados exatamente como queremos.

Alguns comandos úteis:
- `WHERE`: Filtra registros.
- `ORDER BY`: Ordena (`ASC` para crescente, `DESC` para decrescente).
- `LIKE`: Busca por padrão de texto (ex: começa com, contém).


```
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Buscar usuários com mais de 25 anos, ordenando pelo nome em ordem alfabética
cursor.execute("SELECT nome, idade FROM usuarios WHERE idade > 25 ORDER BY nome ASC")

print("Usuários com mais de 25 anos:")
for usuario in cursor.fetchall():
    print(usuario)
```

## Buscando por partes de um texto (`LIKE`)


```
# O símbolo '%' significa 'qualquer texto'. '%@email.com' busca tudo que termina com @email.com
cursor.execute("SELECT * FROM usuarios WHERE email LIKE ?", ('%@email.com',))

print("\nUsuários com @email.com:")
for u in cursor.fetchall():
    print(u)

conn.close()
```
